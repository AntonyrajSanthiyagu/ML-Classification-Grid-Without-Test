```python
import pandas as pd
```

```python
dataset=pd.read_csv("CKD.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
dataset["classification_yes"].value_counts()
```

```python
independent=dataset[['age', 'bp', 'al', 'su', 'bgr', 'bu', 'sc', 'sod', 'pot', 'hrmo', 'pcv',
       'wc', 'rc', 'sg_b', 'sg_c', 'sg_d', 'sg_e', 'rbc_normal', 'pc_normal',
       'pcc_present', 'ba_present', 'htn_yes', 'dm_yes', 'cad_yes',
       'appet_yes', 'pe_yes', 'ane_yes']]
independent
```

```python
dependent=dataset[['classification_yes']]
dependent
```

```python
independent.shape
```

```python
dependent.shape
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn. model_selection import GridSearchCV
param_grid = {'priors':[None], 'var_smoothing':[0]} 
grid = GridSearchCV(GaussianNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted')
grid.fit(independent, dependent)
```

```python
result=grid.cv_results_
#print(result)

y_pred = grid.predict(independent)
#print(y_pred)
   
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(dependent, y_pred)
#print(cm)

from sklearn.metrics import classification_report

clf_report = classification_report(dependent, y_pred)

#print(clf_report)
```

```python
from sklearn.metrics import f1_score

f1_macro=f1_score(dependent,y_pred,average='weighted')

print("The f1_macro value for best parameter {}:".format(grid.best_params_),f1_macro)
```

```python
print("The confusion Matrix:\n",cm)
```

```python
print("The report:\n",clf_report)
```

```python
#AUC ROC stands for “Area Under the Curve” of the “Receiver Operating Characteristic” curve. 
from sklearn.metrics import roc_auc_score

roc_auc_score(dependent,grid.predict_proba(independent)[:,1])
```
